```python
#Is there a correlation between age and resting blood pressure?
```

```python
correlation = df['age'].corr(df['trestbps'])
print("Correlation between Age and Resting Blood Pressure (trestbps)",round(correlation,2))
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='trestbps', data=df)
#sns.regplot(x='age', y='trestbps', data=df, scatter=False, color='blue')
plt.title('Scatter Plot of Age vs. Resting Blood Pressure (trestbps)')
plt.xlabel('Age')
plt.ylabel('Resting Blood Pressure (trestbps)')
plt.show()
```

```python
#A correlation of 0.27 between age and resting blood pressure (trestbps) suggests a weak positive relationship between these two variables
```

```python
#How does resting blood pressure vary among patients with heart disease compared to those without?
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='trestbps', data=df)
plt.xticks([0, 1], ['No Heart Disease', 'Heart Disease'])
plt.title('Resting Blood Pressure by Heart Disease Status')
plt.xlabel('Heart Disease Status')
plt.ylabel('Resting Blood Pressure (trestbps)')
plt.show()

# Display summary statistics
resting_bp_summary = df.groupby('target')['trestbps'].describe()
print("Summary of Resting Blood Pressure by Heart Disease Status:\n", resting_bp_summary)
```

```python
#Insights
#The box plots showing that resting blood pressure ranges are fairly similar in both groups, though the maximum values are slightly higher in patients without heart disease (200) compared to those with heart disease (180).
```

```python
# Are higher cholesterol levels associated with a higher chance of heart disease?
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='chol', data=df)
plt.xticks([0, 1], ['No Heart Disease', 'Heart Disease'])
plt.title('Cholesterol Levels by Heart Disease Status')
plt.xlabel('Heart Disease Status')
plt.ylabel('Cholesterol Level (chol)')
plt.show()
```

```python
# Display summary statistics
cholesterol_levels_summary = df.groupby('target')['chol'].describe()
print("Summary of cholesterol levels by Heart Disease Status:\n", cholesterol_levels_summary)
```

```python
# The median (50th percentile) cholesterol level is higher for individuals without heart disease (249) compared to those with heart disease (234). This aligns with the boxplot, where the median line in the "No Heart Disease" group appears higher than in the "Heart Disease" group.
```

```python
#How does the presence of exercise-induced angina (exang) differ between patients with and without heart disease?
```

```python
pd.crosstab(df.target,df.exang)
```

```python
plt.figure(figsize=(10, 6))

# Create a count plot of exercise-induced angina (exang) by heart disease status (target)
sns.countplot(x='target', hue='exang', data=df, palette='viridis')

# Set plot labels and title
plt.xticks([0, 1], ['No Heart Disease', 'Heart Disease'])
plt.title('Presence of Exercise-Induced Angina by Heart Disease Status')
plt.xlabel('Heart Disease Status')
plt.ylabel('Number of Patients')
plt.legend(title='Exercise-Induced Angina', labels=['No Angina', 'Angina'])

# Show the plot
plt.show()
```

```python
# Insights
#The visualization reveals that angina does not always indicate heart disease, as shown by its higher prevalence in 
#patients without heart disease.
```
